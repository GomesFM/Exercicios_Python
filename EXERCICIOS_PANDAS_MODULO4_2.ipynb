Exercicios desenvolvidos durante o quarto modulo do curso de data science com biblioteca Pandas da Alura.

Tratamento e manipulação de dados, projeto locação de imoveis.

```python
import pandas as pd
```

```python
dados_locacao = pd.read_json('https://raw.githubusercontent.com/GomesFM/Exercicios_Python/main/dados_exercicios/dados_locacao_imoveis.json')
dados_locacao
```

```python
dados_locacao = pd.json_normalize(dados_locacao['dados_locacao'])
dados_locacao
```

```python
dados_locacao = dados_locacao.explode(['datas_combinadas_pagamento', 'datas_de_pagamento', 'valor_aluguel'])
dados_locacao
```

```python
dados_locacao.info()
```

```python
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','.').replace('reais','').strip())
dados_locacao
```

```python
dados_locacao.reset_index(inplace=True, drop=True)
```

```python
import numpy as np
```

```python
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].astype(np.float64)
dados_locacao
```

```python
dados_locacao.info()
```

```python
dados_locacao['apartamento'] = dados_locacao['apartamento'].str.replace('(blocoAP)', '')
dados_locacao
```

```python
dados_locacao['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_locacao['datas_de_pagamento'] = pd.to_datetime(dados_locacao['datas_de_pagamento'], format='%d/%m/%Y')
dados_locacao
```

```python
dados_locacao.info()
```

```python
dados_locacao['atraso'] = (dados_locacao['datas_de_pagamento'] - dados_locacao['datas_combinadas_pagamento']).dt.days
dados_locacao
```

```python
media_atraso = dados_locacao.groupby(dados_locacao['apartamento'])['atraso'].mean().sort_values(ascending=False)
media_atraso
```
