```python
import pandas as pd
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados_alunos = pd.read_csv(url)
dados_alunos
```

```python
dados_alunos.fillna(0, inplace=True)
```

```python
dados_alunos.isnull().sum()
```

```python
alunos_fora_disciplina = dados_alunos.query('Nome == "Alice" | Nome == "Carlos"').index
alunos_fora_disciplina
```

```python
dados_alunos.drop(alunos_fora_disciplina, axis=0, inplace=True)
dados_alunos.query('Nome == "Alice" | Nome == "Carlos"').index
```

```python
alunos_aprovados = dados_alunos[dados_alunos['Aprovado'] == True]
alunos_aprovados
```

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
```

```python
alunos_aprovados['Notas'].replace(7.0, 8.0, inplace=True)
alunos_aprovados
```

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
```

```python
dados_alunos['Ponto_extras'] = dados_alunos['Notas'].apply(lambda x: x * 0.4)
dados_alunos
```

```python
dados_alunos['Nota_final'] = dados_alunos['Notas'] + dados_alunos['Ponto_extras']
dados_alunos
```

```python
dados_alunos['Aprovado_final'] = dados_alunos['Nota_final'].apply(lambda x: True if x >= 6 else False)
dados_alunos
```

```python
novos_alunos_aprovados = dados_alunos[(dados_alunos['Aprovado'] == False) & (dados_alunos['Aprovado_final'] == True)]
novos_alunos_aprovados
```

```python
alunos_aprovados_final = pd.concat([alunos_aprovados, novos_alunos_aprovados], axis=0, ignore_index=True)
alunos_aprovados_final
```

```python
alunos_aprovados_final.to_csv('alunos_aprovados_final.csv', index=False)
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados_imoveis = pd.read_csv(url, sep=';')
dados_imoveis
```

```python
dados_apartamentos = dados_imoveis.query('Tipo == "Apartamento"')
dados_apartamentos
```

```python
media_quartos = dados_apartamentos[['Quartos']].mean()
media_quartos
```

```python
len(dados_apartamentos.Bairro.unique())
```

```python
media_aluguel_bairro = dados_apartamentos.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False)
media_aluguel_bairro.head(15)
```

```python
media_aluguel_bairro = dados_apartamentos.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False).head()
media_aluguel_bairro.plot(kind='barh', figsize=(10, 8), color='blue')
```
