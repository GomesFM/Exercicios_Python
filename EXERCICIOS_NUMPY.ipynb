Para a execução dos exercicios foi utilizada o dataset Oranges vs. Grapefruit disponivel no Keggle <https://www.kaggle.com/datasets/joshmcadams/oranges-vs-grapefruit>

```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
```

```python
dados = np.loadtxt(url, delimiter=',', usecols=np.arange(1,6,1), skiprows=1)
dados
```

```python
peso_laranja = dados[0:5000, 0]
diametro_laranja = dados[0:5000, 1]
```

```python
peso_toranja = dados[5000:, 0]
diametro_toranja = dados[5000:, 1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(peso_laranja,diametro_laranja)
```

```python
plt.plot(peso_toranja,diametro_toranja)
```

```python
xl = peso_laranja
yl = diametro_laranja
xt = peso_toranja
yt = diametro_toranja
nl = np.size(peso_laranja)
nt = np.size(peso_toranja)
```

```python
al = (nl * np.sum(xl * yl) - np.sum(xl) * np.sum(yl)) / (nl * np.sum(xl**2) - np.sum(xl)**2)
al
```

```python
bl = np.mean(yl) - al * np.mean(xl)
bl
```

```python
ytl = al * xl + bl
```

```python
plt.plot(peso_laranja,diametro_laranja)
plt.plot(xl,ytl)
```

```python
at = (nt * np.sum(xt * yt) - np.sum(xt) * np.sum(yt)) / (nl * np.sum(xt**2) - np.sum(xt)**2)
at
```

```python
bt = np.mean(yt) - at * np.mean(xt)
bt
```

```python
ytt = at * xt + bt
```

```python
plt.plot(peso_toranja,diametro_toranja)
plt.plot(xt,ytt)
```

```python
np.random.seed(13)
coef_angulares_laranja = np.random.uniform(low=1, high=99, size=100)
norma_laranja = np.array([])
for i in range(100):
  norma_laranja = np.append(norma_laranja, np.linalg.norm(yl - ((coef_angulares_laranja[i]*xl)+27)))
```

```python
coef_angulares_laranja
```

```python
norma_laranja
```

```python
j = 1000000
indice = 0
for i in range(100):
  if norma_laranja[i] < j:
    j = norma_laranja[i]
    indice = i
    print(indice)

coef_angulares_laranja[indice]
```

```python
plt.plot(peso_laranja,diametro_laranja)
plt.plot(xl,(coef_angulares_laranja[89]*xl + 17))
```
