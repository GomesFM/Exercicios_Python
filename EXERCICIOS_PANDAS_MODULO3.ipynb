```python
dados_gee = pd.read_excel('/content/drive/MyDrive/ALURA/1-SEEG10_GERAL-BR_UF_2022.10.27-FINAL-SITE.xlsx', sheet_name='GEE Estados')
dados_gee
```

```python
estados_sul = dados_gee[(dados_gee['Estado'] == 'RS') | (dados_gee['Estado'] == 'SC') | (dados_gee['Estado'] == 'PR')]
estados_sul['Estado'].unique()
```

```python
uso_terra_amazonas = dados_gee[(dados_gee['Nível 1 - Setor'] == 'Mudança de Uso da Terra e Floresta') & (dados_gee['Estado'] == 'AM')]
uso_terra_amazonas
```

```python
dados_gee.loc[(dados_gee['Nível 1 - Setor'] == 'Agropecuária') & (dados_gee['Estado'] == 'PA'), 2021].max()
```

```python
dados_gee_emissao = dados_gee[dados_gee['Emissão / Remoção / Bunker'] == 'Emissão']
dados_gee_emissao
```

```python
dados_gee_emissao.drop(columns = 'Emissão / Remoção / Bunker', inplace=True)
dados_gee_emissao
```

```python
colunas_info = list(dados_gee_emissao.loc[:, 'Nível 1 - Setor':'Produto'].columns)
colunas_emissao = list(dados_gee_emissao.loc[:, 1970:2021].columns)
dados_gee_ano = dados_gee_emissao.melt(id_vars=colunas_info, value_vars=colunas_emissao, var_name='Ano', value_name='Emissão')
dados_gee_ano
```

```python
dados_gee_ano.groupby('Nível 1 - Setor').groups
```

```python
dados_gee_ano.groupby('Nível 1 - Setor').get_group('Agropecuária')
```

```python
dados_gee_ano[dados_gee_ano['Ano'] == 2021].groupby('Nível 1 - Setor')[['Emissão']].mean()
```

```python
dados_gee_ano.groupby('Nível 1 - Setor')[['Emissão']].sum().sort_values('Emissão', ascending=False)
```

```python
dados_gee_setores = dados_gee_ano.pivot_table(values='Emissão', index='Ano', columns='Nível 1 - Setor', aggfunc='mean')
dados_gee_setores
```

```python
dados_gee_setores.plot(subplots=True, figsize=(10,30));
```

```python
dados_gee_estado_setor = dados_gee_ano.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
dados_gee_estado_setor
```

```python
dados_gee_estado_setor.xs('Energia', level=1)
```

```python
dados_gee_estado_setor.xs('MG', level=0).idxmax()
```

```python
dados_gee_estado_setor.groupby(level=0).idxmax()
```
