```python
def identifica_outliers(column, dataframe):
  q1 = dataframe[column].quantile(.25)
  q3 = dataframe[column].quantile(.75)
  iqr = q3 - q1
  limite_inferior = q1 - (1.5 * iqr)
  limite_superior = q3 + (1.5 * iqr)

  outliers = dataframe[(dataframe[column] < limite_inferior) | (dataframe[column] > limite_superior)]
  return outliers
```

```python
print(identifica_outliers('Valor da transação', transacoes))
```
