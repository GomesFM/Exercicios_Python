Exercicios desenvolvidos durante o quarto modulo do curso de data science com biblioteca Pandas da Alura.
Tratamento e manipulação de dados, projeto loja de vendas online.

```python
import pandas as pd
```

```python
dados_vendas = pd.read_json('https://raw.githubusercontent.com/GomesFM/Exercicios_Python/main/dados_exercicios/dados_vendas_clientes.json')
dados_vendas
```

```python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
dados_vendas
```

```python
dados_vendas = dados_vendas.explode(['Cliente', 'Valor da compra'])
dados_vendas
```

```python
dados_vendas.reset_index(inplace=True, drop=True)
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip())
dados_vendas
```

```python
import numpy as np
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].astype(np.float64)
```

```python
dados_vendas.info()
```

```python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower().replace('[^a-z ]', '', regex=True)
dados_vendas
```

```python
dados_vendas['Data de venda'] = pd.to_datetime(dados_vendas['Data de venda'], format='%d/%m/%Y')
```

```python
dados_vendas
```

```python
dados_vendas.info()
```

```python
subset = dados_vendas.groupby(dados_vendas['Cliente'])['Valor da compra'].sum().sort_values(ascending=False)
subset
```
